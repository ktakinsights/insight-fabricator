## Load and filter initial dataset

We begin by loading our initial dataset, renamed and filtered according to the recommendations in our documentation that we noted in the base project.  This filtered version focuses on the 69 DRG groups that have more than 2000 claims each.  This reduces our dataset by about 20%.

```python
# Load necessary libraries
import os
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import MultiComparison
import seaborn as sns
import matplotlib.pyplot as plt

# Get the absolute path of the directory containing the script file
script_dir = os.path.abspath(os.getcwd())

# Run clean_data.py in the same directory
os.system(f"python {os.path.join(script_dir, 'clean_data.py')}")

# Load the cleaned data
data_filepath = os.path.join(script_dir, 'data_cleaned.csv')
data = pd.read_csv(data_filepath)

# Show the first few rows of the cleaned data
print(data.head())
```

# Subgroup Analysis
Subgroup analysis focuses on specific notable DRGs or procedure codes. We investigate how factors such as age and gender impact healthcare costs within these subgroups.  We aim to identify trends or variations that may not be obvious in the broader dataset.

## Selection of DRGs
For simplicity, we focus on DRG codes associated with high costs, occurrences, or cost variability.  Note that the analysis could be performed by ICD procedure code instead â€“ the process is functionally the same.  We choose a cutoff threshold of the top 5%.

Highest total cost and highest average cost: Looking at DRGs with the highest total or average cost helps us focus on the biggest sources of healthcare spending.  

High frequency of occurrence: When we study DRGs that often happen, we can learn more about common medical conditions or procedures that affect many people.

High variability of cost: We define measure variability with coefficient of variation(CV). Exploring DRGs or procedure codes with high CV shows us where there are big differences in healthcare spending among patients. We can look into the reasons for these differences, such as varying treatment methods, patient backgrounds, or healthcare provider practices.

### List of codes to analyze
We define the top 5% for each metric, resulting in the identification of 4 DRGs per metric. As we have selected four distinct metrics, we obtain four separate lists, each containing four DRGs. Among these lists, two DRGs appear on two different lists, reducing our focus from the original 69 base DRG codes to a more manageable 14.

To better understand the results obtained so far, let's explain the meaning of the lists:

Total cost list: The entries in this list represent the total amount spent (in dollars, $) for each DRG.  Multiply by 20 to estimate the total cost of the population.

Average cost list: The entries in this list represent the average cost (in dollars, $) of each claim within a particular DRG.

Frequency list: The entries in this list show the total number of claims within each DRG.  Multiply by 20 to estimate the total cost of the population.

Coefficient of variation list: The entries in this list provide information about the spread, indicating the extent of variability in relation to the mean of the data set.

```python
# Define a cutoff - 5%
cutoff = int(np.ceil(len(data['base_drg'].unique()) * 0.05))

# Print the cutoff value
print("Cutoff value: ", cutoff)

# Total cost list
total_cost_list = data.groupby('base_drg')['payment'].sum()
sg_tc_list = total_cost_list.sort_values(ascending=False).head(cutoff)
print('Total Cost List:\n', sg_tc_list)

# Frequency list
frequency_list = data.groupby('base_drg').size()
sg_fr_list = frequency_list.sort_values(ascending=False).head(cutoff)
print('Frequency List:\n', sg_fr_list)

# Average cost list
average_cost_list = total_cost_list / frequency_list
sg_ac_list = average_cost_list.sort_values(ascending=False).head(cutoff)
print('Average Cost List:\n', sg_ac_list)

# Coefficient of variability list
cv_list = data.groupby('base_drg')['payment'].std() / average_cost_list
sg_cv_list = cv_list.sort_values(ascending=False).head(cutoff)
print('CV List:\n', sg_cv_list)


# Filter dataset to only contain the base_drgs in the four lists
sg_data = data[data['base_drg'].isin(list(sg_tc_list.index)
                                     + list(sg_fr_list.index)
                                     + list(sg_ac_list.index)
                                     + list(sg_cv_list.index))]
```

##  Impacts of age and gender

### Gender
We conduct independent samples t-tests to compare average payments between male and female patients within each base DRG. We choose drgs resulting from tests that have p < .05.  These base DRGs should be further investigated to understand the reasons for the observed differences in payments by gender.

```python
# t-test and print results
gender_base_drgs = []

for base_drg in sg_data['base_drg'].unique():
    subset = sg_data[sg_data['base_drg'] == base_drg]
    male_payments = subset[subset['gender'] == 'Male']['payment']
    female_payments = subset[subset['gender'] == 'Female']['payment']
    t_stat, p_value = ttest_ind(male_payments, female_payments)
    if p_value < 0.05:
        gender_base_drgs.append((base_drg, subset['descriptor_drg'].iloc[0]))
        print(f"Base DRG: {base_drg} | t-statistic: {t_stat:.3f} \
              |p-value: {p_value:.3f}")
```

#### Visualizing Payment Differences by Gender in specific DRGs

```python
# Create a DataFrame with only significant base DRGs
significant_base_drgs_df = sg_data[
    sg_data['base_drg'].isin([drg[0] for drg in gender_base_drgs])]

# Set up plot aesthetics
sns.set(style="whitegrid")
palette = sns.color_palette("muted")

# Create a catplot with violin plots and separate y-axes for each gender
g = sns.catplot(
    x="gender",
    y="payment",
    col="descriptor_drg",
    col_wrap=2,
    data=significant_base_drgs_df,
    kind="violin",
    order=["Male", "Female"],
    palette=palette,
    scale="width",
    inner="quartile",
    sharey=False,
    height=6,
    aspect=1.2
)

# Set axis labels and titles
g.set_axis_labels("Gender", "Payment")
g.set_titles(col_template="{col_name}")

# Customize plot appearance
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle(
    "Payment Distribution by Gender for Significant Base DRGs", fontsize=16)

# Display the plot
plt.show()
```

### Age
We performed one-way ANOVA tests to assess the impact of age on payments within each base DRG. For the base DRGs with a p-value less than 0.05, we observe a statistically significant difference in payments across age groups. To further examine these differences, we conduct pairwise comparisons using the Tukey HSD test. The base DRGs with at least one significant pairwise comparison are identified as "Age-significant DRGs." Further investigation is warranted to understand the underlying reasons for the observed differences in payments by age within these groups.

### Sample Age ANOVA

```python
# One-way ANOVA for age within each base_drg
age_significant_drgs = []
anova_significant_drgs = []

# Initialize a variable to control the printing of tables
first_significant_found = False

# Loop through unique base DRGs
for base_drg in sg_data['base_drg'].unique():
    # Create a subset of data for the current base DRG
    subset = sg_data[sg_data['base_drg'] == base_drg]
    # Define formula for one-way ANOVA with age as the categorical variable
    formula = 'payment ~ C(age)'
    # Fit one-way ANOVA model
    model = ols(formula, data=subset).fit()
    # Calculate ANOVA table
    anova_table = sm.stats.anova_lm(model, typ=1)

    # Get p-value from ANOVA table
    p_value = anova_table['PR(>F)'].iloc[0]
    # Check if p-value is less than 0.05 (i.e., if ANOVA is significant)
    if p_value < 0.05:
        # If ANOVA is significant, add base DRG to list
        anova_significant_drgs.append(base_drg)

        # To save space, we print a single ANOVA table and Tukey HSD table
        if not first_significant_found:
            print(f"Base DRG: {base_drg}\n{anova_table}")
            # Perform Tukey HSD test on payment by age group
            mc = MultiComparison(subset['payment'], subset['age'])
            tukey_result = mc.tukeyhsd()
            print(tukey_result)
            first_significant_found = True

        # Check if any pairwise age differences are significant
        if tukey_result.reject.any():
            # If any pairwise differences are significant, add to list
            age_significant_drgs.append(base_drg)

print("Anova significant DRGs:", anova_significant_drgs)
print("Age-significant DRGs:", age_significant_drgs)
```

### Age - gender pairs

This analysis echoes the one-way ANOVA testing performed on age categories. However, the two-way ANOVA goes a step further by identifying potentially noteworthy pairs to investigate, particularly focusing on significant interactions between age and gender. As with the previous approach, we use Tukey's HSD post-hoc to mitigate the risk of false positives.

```python
# Two-way ANOVA for age-gender pairs
for base_drg in sg_data['base_drg'].unique():
    subset = sg_data[sg_data['base_drg'] == base_drg]
    formula = 'payment ~ C(age) + C(gender) + C(age):C(gender)'
    model = ols(formula, data=subset).fit()
    anova_table = sm.stats.anova_lm(model, typ=1)

# Tukey HSD for two-way ANOVA (age and gender)
for base_drg in sg_data['base_drg'].unique():
    # Create a copy of the subset
    subset = sg_data[sg_data['base_drg'] == base_drg].copy()
    subset['age_gender'] = subset['age'].astype(str) + '_' + subset['gender']
    mc = MultiComparison(subset['payment'], subset['age_gender'])
    tukey_result = mc.tukeyhsd()
```

```python
# Initialize list for significant base DRGs
pair_anova_base_drg = []

# Two-way ANOVA for age-gender pairs
for base_drg in sg_data['base_drg'].unique():
    subset = sg_data[sg_data['base_drg'] == base_drg]
    formula = 'payment ~ C(age) + C(gender) + C(age):C(gender)'
    model = sm.formula.ols(formula, data=subset).fit()
    anova_table = sm.stats.anova_lm(model, typ=1)

    # Check if p-value for age, gender, or interaction terms are less than 0.05
    if any(anova_table['PR(>F)'][:3] < 0.05):
        pair_anova_base_drg.append(base_drg)

# Print list of significant base DRGsthe
print("Base DRGs with significant age, gender, or interaction effects:")    
print(pair_anova_base_drg)

# Initialize list for significant (base_drg, age_gender) pairs
significant_pairs = []

# Two-way ANOVA for age-gender pairs and Tukey's HSD test for selected base DRGs
for base_drg in sg_data['base_drg'].unique():
    if base_drg in pair_anova_base_drg:
        subset = sg_data[sg_data['base_drg'] == base_drg].copy()  # Create a copy of the subset
        subset['age_gender'] = subset['age'].astype(str) + '_' + subset['gender']
        mc = MultiComparison(subset['payment'], subset['age_gender'])
        tukey_result = mc.tukeyhsd()
        
        # Append significant (base_drg, age_gender) pairs to list
        for i in range(len(tukey_result._results_table.data)):
            if tukey_result._results_table.data[i][-1] == True:
                significant_pairs.append((base_drg, tukey_result._results_table.data[i][0]))

# Print list of significant (base_drg, age_gender) pairs
print("Snapshot of Significant (base_drg, age_gender) pairs:")
print('There are',len(set(significant_pairs)),'total pairs to investigate')
print(set(significant_pairs))
```

# Conclusion

## Findings 
Our analysis focused on the top 5% of Diagnosis Related Groups (DRGs) based on total cost, average cost, count, and coefficient of variation. This led to the identification of 14 unique DRGs, with two DRGs appearing in more than one category.

We found statistically significant differences in payments across genders in half of these DRGs (7 out of 14) via independent t-tests. Age-related differences were detected in 11 DRGs through one-way ANOVA, a finding further substantiated by post hoc Tukey's HSD test.

Further, the interaction between age and gender was also explored through a two-way ANOVA and Tukey's HSD test. This approach unearthed 51 significant unique age-gender pairs across 13 DRGs, indicating potential areas for deeper investigation.

## Limitations
It's important to bear in mind the constraints of this study. Primarily, this is an exploratory analysis, and the results should not be used to confirm research hypotheses. Instead, they help identify areas that might need more focused research, to be conducted on independent, external datasets.

Secondly, the approach used here is susceptible to type I errors, or false positives. Although we've tried to mitigate this risk using Tukey's HSD test, it's likely that our list still contains some false positives. Therefore, it's crucial to treat these findings as starting points rather than conclusive evidence.

Lastly, computational resources could be strained by extensive investigations based on these findings. Hence, we recommend engaging domain experts to further refine this list, ensuring that future investigations are both efficient and meaningful.
